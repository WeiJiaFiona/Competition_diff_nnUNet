计算球形度

```python
import SimpleITK as sitk
import matplotlib.pyplot as plt
import os
from utils import RadiomicsShape, DrawSavePNG
import seaborn as sns
import numpy as np
```

```python
SphericityList=[]

img_small_tumor_path = r"C:\Users\DELL\Desktop\dataset\new_dataset\Dataset002_smalltumor\imagesTr"
gt_small_tumor_path = r"C:\Users\DELL\Desktop\dataset\new_dataset\Dataset002_smalltumor\labelsTr"

img_name = os.listdir(img_small_tumor_path)
gt_name = os.listdir(gt_small_tumor_path)

for i in range(len(img_name)):
    image_path = os.path.join(img_small_tumor_path,img_name[i])
    label_image_path = os.path.join(gt_small_tumor_path ,gt_name[i])
    image = sitk.ReadImage(image_path)
    mask = sitk.ReadImage(label_image_path)
    obj = RadiomicsShape(image,mask)
    SphericityList.append(obj.getSphericityFeatureValue())

DrawSavePNG(SphericityList,"Sphericity")
```

```python
print(f"the mean spherity of {len(img_name)} small tumors is {np.mean(SphericityList)}")
print(f"the variance spherity of {len(img_name)} small tumors is {np.var(SphericityList)}")
```
